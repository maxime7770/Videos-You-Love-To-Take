```python
import tensorflow as tf
import pandas as pd, numpy as np
from IPython.display import YouTubeVideo
```

```python
record = "../data/yt8m/frame/train0208.tfrecord"
# record = "../data/yt8m/frame/validate0638.tfrecord"
# record = "../data/yt8m/frame/test0397.tfrecord"

# Open the TFRecord file
raw_dataset = tf.data.TFRecordDataset(record)
raw_dataset
```

```python
# Read 1 example
for serialized_example in raw_dataset.take(1):  # Only inspecting the first record for demonstration
    # Parse the SequenceExample
    sequence_example = tf.train.SequenceExample.FromString(serialized_example.numpy())
    
    # Print summary
    print("List of context features:")
    context = sequence_example.context.feature
    for key, value in context.items():
        print(key)
    print("\nList of sequence features:")
    sequences = sequence_example.feature_lists.feature_list
    for key, value in sequences.items():
        print(key)

    # Iterate through context features
    context = sequence_example.context.feature
    print("\nContext Features:")
    for key, value in context.items():
        print(key, value)
    
    # Iterate through sequence features
    sequences = sequence_example.feature_lists.feature_list
    print("Sequence Features:")
    for key, value in sequences.items():
        print(key)
        for item in value.feature:
            print("  ", item)
```

```python
# Get number of videos in the dataset
num_elements = sum(1 for _ in raw_dataset)
print("\nNumber of videos in the dataset:", num_elements)
```
